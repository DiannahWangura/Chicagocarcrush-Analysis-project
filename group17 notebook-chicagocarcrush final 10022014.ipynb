Stakeholder:
The primary stakeholder in this scenario is the Vehicle Safety Board of Chicago. They are launching a new campaign with the goal of reducing car crashes in the city.

Business Problem:
The business problem revolves around determining which car crashes are preventable and which are not. Specifically:
Preventable Crashes: These are accidents that could have easily been avoided. They typically result from factors such as not following traffic laws or negligent driving.

Less Preventable Crashes: These incidents would require substantial resources (money, time, and labor) to address. Examples include bad road conditions, vision obscurity, and adverse weather conditions.
The Vehicle Safety Board has the capacity to focus on large-scale projects, so our focus will be on the “Less Preventable” crashes that are not directly related to human error1.
Additionally, we aim to target specific age groups most involved in crashes to tailor our prevention efforts effectively
